##  <mark><span  style= "font-family: Times New Roman">Import necessary libraries</span> </mark> <a class="anchor" id="Import_necessary_libraries"></a>

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

##  <mark><span  style= "font-family: Times New Roman">View the data</span> </mark> <a class="anchor" id="Import_necessary_libraries"></a>

```python
# Load the data
data = {'Batter 1 Score': [56, 19, 0, 68, 16, 67, 85, 101, 78],
        'Batter 2 Score': [60, 58, 38, 71, 69, 66, 68, 41, 67],
        'Temperature': [31, 33, 30, 35, 34, 15, 10, 7, 11]}

df = pd.DataFrame(data)

# Display the data
print(df)
```


##  <mark><span  style= "font-family: Times New Roman">Data preprocessing</span> </mark> <a class="anchor" id="Import_necessary_libraries"></a>

<div class="warning" style='padding:0.1em; background-color:#E9D8FD; color:#69337A'>
<span>
<p style='margin-top:1em; text-align:left'>
</p>
<p style='margin-left:1em;'>
</p><br>
   
  1. Third match score of Batsman-1 was mistakenly noted as 0. Kindly replace the value with the mean of earlier matches as it is not known.Round the calculated mean to the nearest integer. <br>    

2. Create a new feature in the dataframe naming 'weather' <br>
    
   a. Column should have 'Warm' if the temperature on the match day is greater than the mean of the temperatures.  <br>
   b. Column should have 'Cool' if the temperature on the match day is lesser than the mean of the temperatures.  <br>
</span>
</div>


```python
# Data preprocessing

# Replace the incorrect value in 'Batter 1 Score' with the mean of earlier matches
# (excluding the incorrect value)
corrected_value = df['Batter 1 Score'][:-1].mean().round()
df.loc[2, 'Batter 1 Score'] = corrected_value

# Create a new feature 'weather'
mean_temperature = df['Temperature'].mean()

# Column should have 'Warm' if the temperature on the match day is greater than the mean
# Column should have 'Cool' if the temperature on the match day is lesser than the mean
df['Weather'] = ['Warm' if temp > mean_temperature else 'Cool' for temp in df['Temperature']]

print(df)

```

##  <mark><span  style= "font-family: Times New Roman">Perform EDA</span> </mark> <a class="anchor" id="Import_necessary_libraries"></a>

### Visualize the evidences 

```python
# EDA and Visualization

# Visualize Batter 1 Score and Batter 2 Score in different weather conditions

# Set the style of seaborn
sns.set(style="whitegrid")

# Create a figure with subplots
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 5))

# Plot for Batter 1 Score
sns.boxplot(x='Weather', y='Batter 1 Score', data=df, ax=axes[0])
axes[0].set_title('Batter 1 Score in Different Weather Conditions')

# Plot for Batter 2 Score
sns.boxplot(x='Weather', y='Batter 2 Score', data=df, ax=axes[1])
axes[1].set_title('Batter 2 Score in Different Weather Conditions')

# Show the plots
plt.show()
```

```python
# Calculate the mean scores for each batter in different weather conditions as required
mean_scores = df.groupby('Weather').mean()

# Displaying the mean scores
print(mean_scores)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'df' is the DataFrame containing the data
sns.boxplot(x='Weather', y='Batter 1 Score', data=df, showmeans=True, meanline=True, meanprops={"marker":"o", "markerfacecolor":"white", "markeredgecolor":"black"})
plt.title('Batter 1 Score in Different Weather Conditions with Mean')
plt.show()

sns.boxplot(x='Weather', y='Batter 2 Score', data=df, showmeans=True, meanline=True, meanprops={"marker":"o", "markerfacecolor":"white", "markeredgecolor":"black"})
plt.title('Batter 2 Score in Different Weather Conditions with Mean')
plt.show()
```
